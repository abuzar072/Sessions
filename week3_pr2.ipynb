```python
from google.colab import userdata
import os

# Get API key from Colab secret
api_key = userdata.get("GROQ_API")

# Save into environment variable
os.environ["GROQ_API_KEY"] = api_key

print("API key loaded successfully")
```

```python
from groq import Groq

client = Groq(
    api_key=os.environ.get("GROQ_API_KEY")
)

response = client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=[
        {
            "role": "user",
            "content": "Hello, who are you?"
        }
    ]
)

print(response.choices[0].message.content)
```

```python
def chat_with_bot(message, history):
    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[
            {"role": "system", "content": "You are a helpful AI assistant."},
            {"role": "user", "content": message}
        ]
    )

    return response.choices[0].message.content
```

```python
import gradio as gr

# Custom Theme
theme = gr.themes.Base(
    primary_hue="purple",
    secondary_hue="pink",
    neutral_hue="slate",
).set(
    body_background_fill="#0f172a",   # dark blue/black
    body_text_color="#e5e7eb",
    block_background_fill="#111827",
    block_border_width="0px",
    block_shadow="0 0 20px rgba(168,85,247,0.3)",
    button_primary_background_fill="#a855f7",
    button_primary_text_color="white"
)

# Custom CSS for GenZ vibe
custom_css = """
body {
    background: linear-gradient(135deg, #0f172a, #1e1b4b, #312e81);
}

.gradio-container {
    font-family: 'Poppins', sans-serif;
}

.message.user {
    background: linear-gradient(135deg, #a855f7, #ec4899);
    color: white;
    border-radius: 15px;
}

.message.bot {
    background: #1f2937;
    color: #e5e7eb;
    border-radius: 15px;
}

footer {
    display: none;
}
"""

# UI
demo = gr.ChatInterface(
    fn=chat_with_bot,
    title="⚡ GenZ AI Chat",
    description="🚀 Fast • Smart • Stylish (Powered by Groq)",
    theme=theme,
    css=custom_css
)

demo.launch(share=True)
```
